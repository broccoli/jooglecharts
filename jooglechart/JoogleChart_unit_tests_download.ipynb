```python
from jooglechart_api import JoogleChart, Filter, SeriesFilter, ChartRow
from jooglechart.utils import get_df
import pandas as pd
# from jooglechart_api.widgets import Box
```

# Chart with all columns and rows visible

```python
df = get_df(cols = 6, rows=4)
chart = JoogleChart(df, chart_type="Table")
chart.add_download()
chart.show()
```

# Chart with hidden column and role column

```python
df = get_df(cols = 6, rows=4)
chart = JoogleChart(df, chart_type="Table")
chart.set_view_cols([0, 2, 3, 4, 5, 6])
chart.set_role(3, "blah")
chart.add_download(position="top")
chart.show()
```

# Chart with native filter and Series Filter

```python
df = get_df(cols = 6, rows=4, category_column=True)
chart = JoogleChart(df, chart_type="ColumnChart")
chart.set_view_cols([0, 2, 3, 4, 5, 6])
chart.add_download(position="top")

f = Filter("CategoryFilter")
f.add_options(filterColumnIndex=1)

chart.add_filter(f)

sf = SeriesFilter()
chart.add_filter(sf)

chart.show()
```

# Chart with sonar row filter and sonar column filter

```python
df = get_df(cols = 6, rows=4, category_column=True)
chart = JoogleChart(df, chart_type="ColumnChart")
chart.set_view_cols([0, 2, 3, 4, 5, 6])
chart.add_download(position="top")

row_filter = Filter("CategoryFilter", data=df.cities)
row_filter.add_options(filterColumnIndex=0)
row_filter.add_sender(key="QWE")


column_filter = Filter("CategoryFilter", data=pd.Series(df.columns.tolist()[2:]))
column_filter.add_options(filterColumnIndex=0)
column_filter.add_sender("ABC")

chart.add_receiver(key="QWE", column=0)
chart.add_receiver(key="ABC", action="filter_columns")

row = ChartRow(row_filter, column_filter, chart)

row.show()
# print row.render()
```
